### 10) WAP to print GCD of given two numbers.

```python
num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))

min = min(num1, num2)
max = max(num1, num2)

rem = max % min
while rem != 0:
    max = min
    min = rem
    rem = max % min

print(min)
```

### 12) WAP to find whether the given number is Prime or not.

```python
num = int(input("Enter the number : "))
flag = True

for i in range(2, (num//2)+1):
    if(num % i == 0):
        flag = False
        break

if flag:
    print(num, "is prime number")
else:
    print(num, "is not prime number")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
num = int(input("Enter the number : "))
cpy = num

digit = 0
while cpy != 0:
    digit += 1
    cpy //= 10

cpy = num
ans = 0

while cpy != 0:
    prod = 1
    for i in range(digit):
        prod *= (cpy % 10)
    ans += prod
    cpy //= 10

if num == ans:
    print(num, "is armstrong")
else:
    print(num, "is not armstrong")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
num = int(input("Enter the nunmber : "))

for i in range(1, num+1):
    ans = 0
    for j in range(1, i//2 + 1):
        if i%j == 0:
            ans += j

    if i == ans:
        print(i)
```
