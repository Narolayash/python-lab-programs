### Practical 1: Count Lines, Words and Characters

**Problem Definition:**  
Write a Python program to read text from a file and count the total number of lines, words, and characters. The result should be written to an output file.


```python
with open("Hello.txt", "r") as fs:
    lines = len(fs.readlines())
    fs.seek(0, 0)
    word = len(fs.read().split())
    fs.seek(0, 0)
    character = len(fs.read())

with open("output.txt", "a") as fs:
    fs.write(f"\nPractical 1: {lines}  {word}  {character}")
```


### Practical 2: Search a Word in a File

**Problem Definition:**  
Write a Python program to search a given word in a text file and count its occurrences. Store the result in an output file.


```python
with open("Hello.txt", "r") as fp:
    search = input("Which word you want to find:")
    words = fp.read().split()
    pure_words = [i.strip("!,.?").lower() for i in words]
    count = pure_words.count(search)

with open("output.txt", "a") as fs:
    fs.write(f"\nPractical 2: word: {search}, search count = {count}")
```


### Practical 3: Student Marks and Average

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Store marks in a list, calculate the average using a user-defined function, and write student details along with average marks to an output file.



```python
def calculate_arg(marks):
    return sum([int(i) for i in marks]) / len(marks)

with open("studentDetails.txt", "r") as fp:
    data = fp.read().split(", ")
    roll, name, marks = data[0], data[1], data[2].split()
    avg = calculate_arg(marks)

with open("output.txt", "a") as fs:
    fs.write(f"\nPractical 3:\nStudent Roll NO: {roll}\nStudent Name: {name}\nStudent marks: {marks} and average of marks: {avg}")
```


### Practical 4: Maximum and Minimum using Tuple

**Problem Definition:**  
Write a Python program to read numbers from a file, store them in a tuple, and find the maximum and minimum values. Write the result to an output file.


```python
with open("numbers.txt", "r") as fp:
    nums = tuple([int(i) for i in fp.read().split()])
    maxx = max(nums)
    minn = min(nums)
    
with open("output.txt", "a") as fs:
    fs.write(f"\nPractical 4: max number: {maxx}\tmin number: {minn}")
```


### Practical 5: Remove Duplicate Words using Set

**Problem Definition:**  
Write a Python program to read words from a file and remove duplicate words using a set. Store the unique words in an output file.


```python
with open("Hello.txt", "r") as fs:
    unique = set([i.strip("!,.?") for i in fs.read().split()])

with open("output.txt", "a") as fs:
    fs.write(f"\nPractical 5: Unique words: { ", ".join(unique) }")
```


### Practical 6: Word Frequency using Dictionary

**Problem Definition:**  
Write a Python program to count the frequency of each word in a text file using a dictionary and write the result to an output file.


```python
with open("Hello.txt", "r") as fs:
    pure_words = [i.strip("!,.!?") for i in fs.read().split()]
    dictt = {}
    for k in pure_words:
        dictt[k.lower()] = dictt.get(k.lower(), 0) + 1

with open("output.txt", "a") as fs:
    fs.write(f"\nPractical 6:result: {dictt}")
```


### Practical 7: Student Result and Grade Calculation

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Use a user-defined function to determine grade (A/B/C/Fail) and write the student result to an output file.

**Grade Criteria:**  
- Marks ≥ 75 : Grade A  
- Marks ≥ 60 : Grade B  
- Marks ≥ 40 : Grade C  
- Marks < 40 : Fail


```python
def cal_grede(marks):
    avg = sum(marks) / len(marks)
    
    if avg >= 75:
        return 'A'
    elif avg >= 60:
        return 'B'
    elif avg >= 40:
        return 'C'
    else:
        return 'Fail'

with open("studentDetails.txt", "r") as fp:
    data = [tuple(i.split(", ")) for i in fp.read().split("\n")]
    modified_data = []
    for i in data:
        i += tuple(cal_grede([int(j) for j in i[2].split()]))
        modified_data.append(i)

    print(modified_data)
    

with open("output.txt", "a") as fs:
    fs.write(f"\nPractical 7:\n {modified_data}")
```
