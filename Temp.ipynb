```python
n = int(input())
list = []

for i in range(n):
    s = input().split()
    match s[0]:
        case 'insert':
            list.insert(s[1], s[2])
        case 'print':
            print(list)
        case 'remove':
            list.remove(s[1])
        case 'append':
            list.append(s[1])
        case 'sort':
            list.sort()
        case 'pop':
            list.pop()
        case 'reverse':
            list.reverse()
        
```
