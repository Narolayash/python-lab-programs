```python
x, y, z, n = input().split();
x = int(x)
y = int(y)
z = int(z)
n = int(n)

list = []

for i in range(x+1):
    for j in range(y+1):
        for k in range(z+1):
            if(i + j + k) != n:
                list.append([i, j, k])

print(list)
```

```python
n = int(input())
list = []

for i in range(n):
    s = input().split()
    match s[0]:
        case 'insert':
            list.insert(s[1], s[2])
        case 'print':
            print(list)
        case 'remove':
            list.remove(s[1])
        case 'append':
            list.append(s[1])
        case 'sort':
            list.sort()
        case 'pop':
            list.pop()
        case 'reverse':
            list.reverse()
        
```
