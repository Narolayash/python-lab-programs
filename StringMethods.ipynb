# capitalize()

The capitalize() method returns a string where the first character is upper case, and the rest is lower case.


```python
txt = "hello, and welcome to my world."

x = txt.capitalize()

print (x)
```

```python
txt = "python is FUN!"

x = txt.capitalize()

print (x)
```

```python
txt = "36 is my age."

x = txt.capitalize()

print (x)
```

# casefold()

```python
txt = "Hello, And Welcome To My World!"

x = txt.casefold()

print(x)
```

# center()

```python
txt = "banana"

x = txt.center(20)

print(x)
```

```python
txt = "banana"

x = txt.center(20, "O")

print(x)
```

```python
txt = "banna"

x = txt.center(20, "O")

print(x)
```
