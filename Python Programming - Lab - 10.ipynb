### 04) WAP to find the longest word(s) in a file.

```python
with open("Hello.txt") as fp:
    data = fp.read().split()
    len_of_longest = len(max(data, key=len))

    print([w for w in data if len_of_longest == len(w)])
```

### 08) WAP to create an UDF named frequency to count occurances of the specific word in a given text file.

```python
def ferquency(word, file):
    with open(file, "r") as fp:
        return (fp.read().lower()).count(word.lower())


ferquency("Programming", "Hello.txt")
```

### 10) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

with open("primenumber.txt", "w") as fp:
    count = 0
    num = 2
    while count < 100:
        if is_prime(num):
            fp.write(str(num) + '\n')
            count += 1
        num +=1 

    
```

```python
def is_prime(n):
    return n > 1 and all(n % i != 0 for i in range(2, n))

with open("primenumbers.txt", "w") as f:
    n = 2
    c = 0
    while c < 100:
        if is_prime(n):
            f.write(f"{n}\n")
            c += 1
        n += 1
```

```python
with open("primenumbers.txt", "w") as fp:
    count = 0
    num = 2

    while count < 100:
        if is_prime(num):
            fp.write(str(num) + "\n")   # each on new line
            count += 1
        num += 1

print("First 100 prime numbers written to file.")
```
