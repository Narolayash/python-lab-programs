### 06) WAP to generate 100 random lottery tickets and pick two lucky tickets from it and announce them as Winner and Runner up respectively.

```python
random.seed(2)
lottery_numbers = [random.randint(10001, 100000) for i in range(100)]

winner, runnerUp = random.sample(lottery_numbers, 2)

print("winner =", winner, "runnerUp =", runnerUp)
```
