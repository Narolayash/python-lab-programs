### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
str = input("Enter the string : ")
Uppercase = Lowercase = Digits = Special = 0

for i in str:
    if i.isupper():
        Uppercase += 1
    elif i.islower():
        Lowercase += 1
    elif i.isdigit():
        Digits += 1
    else:
        Special += 1

print("Uppercase:", Uppercase, "Lowercase:", Lowercase, "Digits:", Digits, "Special:", Special)
```

### 2) Reverse Each Word
- Input: `Python is Easy` → `nohtyP si ysaE`
- Input: `Hello World` → `olleH dlroW`
- Input: `Learn Python Fast` → `nraeL nohtyP tsaF`

```python
str = input("Enter the String :")
li = str.split(" ")
rev = []

for i in li:
    rev.append(i[::-1])

print(" ".join(rev))
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s = input("Enter the string: ")

compression = ''
count = 1

for i in range(1, len(s)):
    if s[i] == s[i-1]:
        count += 1
    else:
        compression += s[i-1] + str(count)
        count = 1

compression += s[-1] + str(count)

print("Compression:", compression)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s = input("Enter the string: ")
toggle = ''

for i in s:
    if i.isupper():
        toggle += i.lower()
    elif i.islower:
        toggle += i.upper()

print(toggle)
```

### 11) Username Validation
#### valid username rules:
1. only alphabets and digits
2. does not start with a digit

- `User123` → Valid

```python
s = input("Enter your username: ")


if not len(s) >= 6:
   print("Invalid")
elif s[0].isdigit():
    print("Invalid")
else:
    for i in s:
        if not i.isalnum():
            print("Invalid")
            break
    else:
        print("valid")
```
