### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
mass = float(input("Enter your weigth (kg):"))
height = float(input("Enter your height (m):"))

def calBMI(mass, height):
    return mass / (height ** 2)

print(round(calBMI(mass, height), 2))
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def isprime(n):
    return 0 if len([i for i in range(2, n//2 + 1) if n%i == 0]) else 1

isprime(17)
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
def ispalindrome(strr):
    return True if strr.lower() == strr[::-1].lower() else False

print(ispalindrome("Yash"))
print(ispalindrome("YaaY"))
print(ispalindrome("Yaay"))
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def fibonacci(n):
    if n == 1:
        return 0
    if n == 2:
        return 1

    return fibonacci(n-1) + fibonacci(n-2)

fibonacci(10)
```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
def lastzerosum(li):
    return sum(i for i in li if i % 10 == 0)

lastzerosum([200, 456, 300, 100, 234, 678])
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invertdict(dictionary):
    return {v: k for k, v in dictionary.items()}
    
dictt = {'a': 10, 'b':20, 'c':30, 'd':40}

invertdict(dictt)
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def uplowcasecounter(strr):
    print("no_upper =", len([i for i in strr if i.isupper()]))
    print("no_lower =", len([i for i in strr if i.islower()]))

uplowcasecounter('AbcDEfgh')
```

### 16) Write udfs to call the functions with following types of arguments:
1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable Legngth Positional(*args) & variable length Keyword Arguments (**kwargs)
5. Keyword-Only & Positional Only Arguments

```python
def posiArguFun(a, b, c):
    print(a, b ,c)

posiArguFun(10, 20, 30)
```

```python
def kwArguFun(a, b, c):
    print(a, b ,c)

kwArguFun(a=10, c=20, b=30)
```

```python
def defaultArguFun(a=10, b=20, c=30):
    print(a, b, c)

defaultArguFun()
defaultArguFun(a=20)
defaultArguFun(b=50)
defaultArguFun(a=40, c=50)
defaultArguFun(a=40, b=80, c=50)
```

```python
def variLenPosiArguFun(*args):
    print(args)

variLenPosiArguFun(10, 20, 30)
```

```python
def variLenKeyWordArguFun(**kwargs):
    print(kwargs)

variLenKeyWordArguFun(a=10, b=20, c=30)
```

```python
def posiOnlyFun(a, b, c, /):
    print(a, b, c)

posiOnlyFun(10, 20, 30)
# posiOnlyFun(10, 20, c=30) TypeError: print() got an unexpected keyword argument 'c'
```

```python
def keyWordOnlyFun(*, a, b, c):
    print(a, b, c)

# keyWordOnlyFun(10, 20, 30) keyWordOnlyFun() takes 0 positional arguments but 3 were given
keyWordOnlyFun(a=10, b=20, c=30)
```
