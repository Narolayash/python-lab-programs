## Practical 1: Student Marks Analyzer

### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks and store result in a dictionary.

### Sample Input
```python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]
```

### Sample Output
```
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

```

````python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]

student_Data = {}

for i in students:
    avg_marks = sum(i[2]) / len(i[2])
    student_Data[i[0]] = (i[1], avg_marks)

print(student_Data)
````

## Practical 2: Inventory Management

### Problem Definition
Remove out-of-stock items and calculate total inventory value.

### Sample Input
```python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
```

### Sample Output
```
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
```

````python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}

update_inventory = {}
total_value = 0

for k, v in inventory.items():
    if v[1] != 0:
        update_inventory[k] = v
        total_value += v[0] * v[1]

print(update_inventory)
print("Total Value:", total_value)
````

````python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 0)
}

update_inventory = {item: (pri, qty) for item, (pri, qty) in inventory.items() if qty > 0}
total_value = sum(pri * qty for pri, qty in update_inventory.values())

print(update_inventory)
print("Total Value:", total_value)
````

## Practical 3: Email Domain Extractor

### Problem Definition
Extract email domains and count occurrences.

### Sample Input
```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
```

### Sample Output
```
{'gmail.com': 2, 'yahoo.com': 1}
```

````python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']

domains = [i[i.find("@")+1:] for i in emails]

freq = {}
for i in domains:
    if i not in freq:
        freq[i] = 1
    else:
        freq[i] += 1

print(freq)
````

````python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']

domains = [i[i.find("@")+1:] for i in emails]

freq = {}
for i in domains:
    freq[i] = freq.get(i, 0) + 1

print(freq)
````

## Practical 4: Shopping Cart Billing

### Problem Definition
Calculate item-wise cost and total bill.

### Sample Input
```
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
```

### Sample Output
```
{'apple': 100, 'banana': 30}
Total Bill: 130
```

````python
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}

bill = {}
total_bill = 0

for product in cart:
    bill[product[0]] = prices[product[0]] * product[1]
    total_bill += prices[product[0]] * product[1]

print(bill)
print("Total Bill:", total_bill)
````

````python
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':100}

bill = {item: prices[item] * qty for item, qty in cart}
total_bill = sum(v for v in bill.values())

print(bill)
print("Total Bill:", total_bill)
````

## Practical 5: Character Position Index

### Problem Definition
Store positions of each character in a string.

### Sample Input
```python
s = "banana"
```

### Sample Output
```
{'b': [0], 'a': [1, 3, 5], 'n': [2, 4]}
```

````python
s = "banana"
posi_char = {}

for i in range(len(s)):
    if s[i] in posi_char:
        posi_char[s[i]].append(i)
    else:
        posi_char[s[i]] = [i]

print(posi_char)

````

## Practical 6: Survey Data Aggregation

### Problem Definition
Count responses city-wise and choice-wise.

### Sample Input
```python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]
```

### Sample Output
```
{'Rajkot': {'Yes': 1, 'No': 1}, 'Ahmedabad': {'Yes': 1}}
```

````python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]

respo_dict = {}

for row in responses:
    if row[1] in respo_dict:
        respo_dict[row[1]][row[2]] = respo_dict[row[1]].get(row[2], 1) + 1
    else:
        respo_dict[row[1]] = {row[2] : 1}

print(respo_dict)
````
