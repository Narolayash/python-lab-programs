### 01) WAP to find their union, intersection, and difference of the given two sets.
**Sample Input:**
```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```
**Sample Output:**
```
Union: {1, 2, 3, 4, 5, 6}
Intersection: {3, 4}
Difference (A-B): {1, 2}

```

````python
A = set([int(i) for i in input("Enter the element: ").split()])
B = set([int(i) for i in input("Enter the element: ").split()])

print("A =", A)
print("B =", B)

print("Union:", A.union(B))
print("Intersection:", A.intersection(B))
print("Difference (A-B):", A.difference(B))
````

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

for k, v in Dict2.items():
    if k in Dict1:
        Dict1[k] += Dict2[k]
    else:
        Dict1[k] = v

print(Dict1)
````

### 11) WAP to find the student with the highest total marks, given student names and their marks.

**Sample Input:**
```
{'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}
```
**Sample Output:**
```
Topper: Amit
```

````python
stu_mark = {'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}
highestMarks = 0
highestMarksKey = ""

for k, v in stu_mark.items():
    sum = 0
    for i in v:
        sum += i
    if sum > highestMarks:
        highestMarks = sum
        highestMarksKey = k

print("Topper:", highestMarksKey)
````

### 12) WAP to extract keys whose values are even numbers.

**Sample Input:**
```
{1: 11, 2: 20, 3: 15, 4: 40}
```
**Sample Output:**
```
[2, 4]
```

````python
dict1 = {1: 11, 2: 20, 3: 15, 4: 40}
li = []

for k, v in dict1.items():
    if v % 2 == 0:
        li.append(k)

print(li)
````

### 15) WAP to create a dictionary using dictionary comprehension from two lists containing student names and marks. Include only those students in the dictionary whose marks are greater than or equal to 75.

**Sample Input:**
```
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]
```
**Sample Output:**
```
{'Amit': 78, 'Neha': 85}
```

````python
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]

student_marks = {k:v for k, v in zip(students, marks) if v >= 75}

print(student_marks)
````
