### Q1:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





```python
heads = int(input("Enter the number of heads : "))
legs = int(input("Enter the number of legs : "))

chicken = (2 * heads) - (0.5 * legs)
dog = heads - chicken

print(f"dogs -> {dog}  chickens -> {chicken}")
```

### Q2:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

```python
n1 = int(input("Enter numerator of first number: "))
d1 = int(input("Enter denominator of first number: "))
n2 = int(input("Enter numerator of second number: "))
d2 = int(input("Enter denominator of second number: "))

n = n1 * d2 + n2 * d1
d = d1 * d2

print(f"you entered fractions: {n1}/{d1} and {n2}/{d2}")
print(f"{n1}/{d1} + {n2}/{d2} = {n}/{d}")
```

### Q3:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.
Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

```python
import math
```

```python
H = int(input("Enter the height of the milk tank: "))
L = int(input("Enter the width of the milk tank: "))
B = int(input("Enter the breadth of the milk tank: "))

h = int(input("Enter the height of the glass: "))
r = int(input("Enter the redias of the glass: "))

print(f"you entered tank dimensions (h,w,b): {H} {L} {B}")
print(f"you entered glass dimensions (h,r): {h} {r}")
print("-----------------------------------------------")
print(f"Volume of tank in cubic cm: {H * L * B}")
print(f"Volume of glass in cubic cm: {math.pi * r * r * h:.2f}")
print(f"{(H * L * B)/(math.pi * r * r * h):.2f} glasses of milk can be obtained")
```
