### 01) WAP to check whether the given number is positive or negative.

```python
num = int(input("Enter the number : "))
if num > 0:
    print(f"The {num} is positive.")
elif num < 0:
    print(f"The {num} is negative.")
else:
    print(f"The {num} is zero.")
```

### 02) WAP to check whether the given number is odd or even.

```python
num = int(input("Enter the number : "))
if num % 2 == 0:
    print(f"The {num} is even.")
else:
    print(f"The {num} is odd.")
```

### 04) WAP to find out largest number from given three numbers.

```python
a, b, c = input("Enter three numbers : ").split()
a = int(a)
b = int(b)
c = int(c)

if a > b and a > c:
    print(f"{a} is largest.")
elif b > c:
    print(f"{b} is largest.")
else:
    print(f"{c} is largest.")
```

### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("Enter the year : "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(f"{year} is leap year")
        else:
            print(f"{year} is not leap year")
    else:
        print(f"{year} is leap year")
else:
    print(f"{year} is not leap year")
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
a, b = input("Enter two numbers : ").split()
a = float(a)
b = float(b)

op = input("Which operation you want to perform (+, -, *, /) : ")

match op:
    case '+': 
        print(f"{a} {op} {b} = {a + b}")
    case '-': 
        print(f"{a} {op} {b} = {a - b}")
    case '*': 
        print(f"{a} {op} {b} = {a * b}")
    case '/': 
        print(f"{a} {op} {b} = {a / b}") if b != 0 else print("divisor is zero")
    case _:
        print("plz enter valid operation")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
m1, m2, m3, m4, m5 = input("Enter marks of five subjects : ").split()
m1 = int(m1)
m2 = int(m2)
m3 = int(m3)
m4 = int(m4)
m5 = int(m5)

percentage = (m1 + m2 + m3 + m4 + m5) / 5;
# print(percentage)

if percentage < 35:
    print(f"{percentage} : Fali")
elif percentage >= 35 and percentage < 45:
    print(f"{percentage} : Pass")
elif percentage >= 45 and percentage < 60:
    print(f"{percentage} : Second class")
elif percentage >= 60 and percentage < 70:
    print(f"{percentage} : First class")
elif percentage >= 70 and percentage <= 100:
    print(f"{percentage} : Distinction")
else:
    print("enter valid marks bro")
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
unit = input("Enter the units: ")
unit = int(unit)

if unit > 0 and unit <= 50:
    bill = unit * 2.60
elif unit > 50 and unit <= 100:
    bill = ((unit-50) * 3.25) + (50 * 2.60) 
elif unit > 100 and unit <= 200:
    bill = ((unit-100) * 5.26) + (50 * 3.25) + (50 * 2.60)
else:
    bill = ((unit-200) * 8.45) + (100 * 5.26) + (50 * 3.25) + (50 * 2.60)

print(F"unit {unit} : bill {bill}")
```
