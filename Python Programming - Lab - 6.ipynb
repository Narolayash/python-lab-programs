### 01) WAP to find sum of all the elements in a List.

```python
li = input("Enter the elements:").split()

sum = 0
for i in li:
    sum += int(i)

print("sum of all the elements in a list", sum)
```

### 05) WAP to split the List into two parts and append the first part to the end.

```python
li = [i for i in input("Enter the elements:").split()]
print("before list:",li)
part1 = li[:len(li)//2]
part2 = li[len(li)//2:]

part2.extend(part1)

print("after list:", part2)
```

### 09) WAP to count unique items in a list.

```python
li = [i for i in input("Enter the elements:").split()]
print("list : ",li)

countUni = 0
for i in li:
    if li.count(i) == 1:
        countUni += 1

print("Number of unique items is", countUni)
```

### 14) WAP to extract elements with frequency greater than K.

```python
li = [i for i in input("Enter the elements:").split()]
k = int(input("Enter the frequency:"))

newli = []
for i in li:
    if li.count(i) > k and i not in newli:
            newli.append(i)

print("Element with frequency greater than k:", newli)      
```

### 15) WAP to create a list of squared numbers from 0 to 9 with and without using List Comprehension.

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
fruits = [i for i in input("Enter the elements:").split()]

bfruit = [i for i in fruits if i[0] == 'b' or i[0] == 'B']
print("bfruit:", bfruit)
```

### 17) WAP to create a list of common elements from given two lists.

```python
li1 = [i for i in input("Enter the elements:").split()]
li2 = [i for i in input("Enter the elements:").split()]

commonli = []
for i in li1:
    if i in li2 and i not in commonli:
        commonli.append(i)

print("Common elements from given two lists are", commonli)
```
